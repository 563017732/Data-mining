```python

import pandas as pa
import numpy as np

np.set_printoptions(suppress=True)           #numpy中数列由科学计数法转化为整数计数
#######
###test######
######
#######
#低于min_support的减枝策略
#######
def min_support_purning(Dict,min_support):
    keys=[]
    for key in Dict:
        if Dict[key]<min_support:
            keys.append(key)
    for i in keys:
        del Dict[i]
    return Dict   

########
#重新组合生成Cn的关键字
########

import math
def generate_Cntitle1(Ln): 
    #############
    #生成新的索引
    ############
    num=len(Ln)
    sumnum=0
    for i in range(num):
        sumnum=i+sumnum
    Lkeys=list(Ln.keys())
    lkeys=["a"]* sumnum
    string=''
    count=0
    for i in range(num-1):
        a=Lkeys[i]
        for j in range(i+1,num):   
            b=Lkeys[j]
            string=a+','+b
            lkeys[count]=string
            count=count+1         
    return lkeys



######################
   #生成字符中切割，删除重复项
###################### 
def Delete_duplicate(n,Cn_title):
    Cn_title2=["a"]* len(Cn_title)
    for i in range(len(Cn_title)): 
        Cn_title2[i]=list(set(Cn_title[i].split(",")))                          #分割重新组合好的item
        Cn_title2[i]=sorted(set(Cn_title2[i]), key = Cn_title2[i].index)        #删除items 中的重复项
        if len(Cn_title2[i])!=n:                                                #删除数量和N 不相等的项
            Cn_title2.pop(i)
    return Cn_title2



########################
  #删除CN-1中不存在的值
######################## 

################################
    #为新的索引统计频率
################################
def generate_numbercode(Cn):
    #把物理编号换成数组编号
    indexnum=np.zeros((len(Cn),len(Cn[0])),dtype = np.int)
    #print(len(Cn),len(Cn[0]))
    for i in range(len(Cn)):
        for j in range(len(Cn[0])):
            for k in index.keys():
                if k==Cn[i][j]:                  
                    indexnum[i][j]=int(index[k])
    #print(indexnum)
    return indexnum
    #对新生成的Cn 进行预剪枝
    
###########################
    #生成新的字典CN
###########################
def genrate_newcn(Cn,numTrasaction,AprioriData,Ln):
    row=Cn.shape[0]          #获取行数
    column=Cn.shape[1]       #获取列数
    #print(row*column)
    nlist=np.zeros((numTrasaction,column),dtype=np.int)
    min=np.zeros(numTrasaction,dtype=np.int)
    Cn_values=np.zeros(row,dtype=np.int)
    for i in range(row):
        for j in range(column):
            for k in range(numTrasaction):
                nlist[k][j]=AprioriData[k][Cn[i][j]]     
        Cn_values[i]=nlist.min(axis=1).sum()                  #获取到合并后的公共购买商品数
    #print(Cn_values)                                    
    Cn_title=generate_Cntitle1(Ln)                             #生成组合后的变量
    Cn_dict=dict(zip(Cn_title,Cn_values))         #生成新的字典Cn
    #print(Cn_dict)
    return Cn_dict
print("**********算法开始************")

Data = pa.read_csv("BreadBasket_DMS.csv")
Data=pa.DataFrame(Data)
print("**********数据集来源kaggle——Transactions from a bakery************")
print(Data.head())
#Data.Transaction         #获取transaction列
#Data.Item.unique()         #获取item 的所有属性值:95

numTrasaction=Data.Transaction.max()      #获取交易笔数:9684
Item=Data.Item.unique()
numItem = len(Data.Item.unique())         #获取item 的所有属性值:95
newData=dict(Data.groupby([Data['Transaction'],Data['Item']]).size())    #按照交易笔数与交易数量对商品进行分组统计

#把物理编号换成数组编号
values=np.arange(numItem)
index=dict(zip(Item, values))
print("***********由数据集构造交易商品编码索引***********")
print(index)

#for key,key2 in newData.keys():                         #实现数据按照交易次数和商品属性分组
#   print("key1=",key,"key2=",key2,"items=",newData[key,key2])
#print(newData.keys())
#aa=(1, 'Breada') in newData         判断是否在数据字典中
#print(Data.Transaction.max() )

############## 
#重新构造数据集
##############
AprioriData =np.zeros((numTrasaction,numItem),dtype=int)         #创造新的数据集，每行表示一个交易，每一列表示购买了一样物品，与购买件数
for i in range(numTrasaction):
    count=0
    for j in Data.Item.unique():
        if (i, j) in newData:
            AprioriData[i-1][count]=newData[i,j]
        count=count+1
#print(AprioriData[0:5,:])
AprioriData.sum( axis= 0)             #统计新数据列的和，即每个属性的次数
print("***********以2维矩阵形式重构数据集AprioriData**********")
print(AprioriData)
########
#构造C1——用字典类型
########
keys=Data.Item.unique()
values=AprioriData.sum( axis= 0)
Cn = dict(zip(keys, values))          #生成字典

#######
#定义最小支持 和最小置信度
#######
min_support=300
min_confidence=0.5

Ln=min_support_purning(Cn,min_support)
count=2
print("L",count-1,"计算结果为：",Ln)
n=len(Ln)
while(n>=1):
    L_temp=Ln
    Cn=generate_Cntitle1(Ln)
    Cn=Delete_duplicate(count,Cn)
    count=count+1
    Cn_number=generate_numbercode(Cn)
    Cn=genrate_newcn(Cn_number,numTrasaction,AprioriData,Ln)
    Ln=min_support_purning(Cn,min_support)
    n=len(Ln)       
    print("**********************")
    print("L",count-1,"计算结果为：",Ln)
    print("**********计算信度************")
    for i in Ln.keys():
        keywords=list(set(i.split(",")))  
        for j in range(len(keywords)):
            key1=keywords[j]
            confidence=Ln[i]/L_temp[key1]
            print(key1,"---->",i,"的信度为：",confidence)
            if confidence>=min_confidence:
                print('\033[1;36m(๑‾ ꇴ ‾๑)这个法则信度好大，关联法则————成立\033[0m')
            else:
                print('\033[0;31m╭(╯^╰)╮世界上不可棱有信度这么小的关联法则————不成立\033[0m')
            print()
print("*********************算法结束*****************") 
```
